```python
data = MqDataset.from_path("./data/MQ2007/", "./cache/")
```

```python
ranker_generator = lambda: LinearRanker(46)
common_params = dict(online_metric=PrivatizedMaxRR(1.0),
                     n_clients=2000,
                     sessions_budget=1000000, 
                     seed=seed,
                     ranker_generator=ranker_generator,
                     antithetic=True,
                     noise_std=1e-2, lr=1e-3)

sessions_per_feedback_grid = [2, 4, 8, 16, 32]
validation = data.folds[0].validation

def get_trajectory(task):
    sessions_per_feedback, click_model = task
    
    params = common_params.copy()
    params.update(dict(sessions_per_feedback=sessions_per_feedback, click_model=click_model))
    train_result = train_uniform(params, validation)
    return train_result.batch_metrics

tasks = itertools.product(sessions_per_feedback_grid, all_click_models)
tasks = list(tasks)
n_cpu = min(80, len(tasks))

with Pool(n_cpu) as p:
    trajectories = p.map(get_trajectory, tasks)
    
click_model2trajectory = dict((m.name, {}) for m in all_click_models)
for (sessions_per_feedback, click_model), results in zip(tasks, trajectories):
    click_model2trajectory[click_model.name][sessions_per_feedback] = results
```

```python
import seaborn as sns
sns.set(style="darkgrid")
from pylab import rcParams
from util import smoothen_trajectory
plt.close('all')
rcParams['figure.figsize'] = 20, 5

f, ax = plt.subplots(1, 3)

for model_id, model_name in enumerate([m.name for m in all_click_models]):
    for i, sessions_per_feedback in enumerate(sessions_per_feedback_grid):
        trajectory = click_model2trajectory[model_name][sessions_per_feedback]
        if sessions_per_feedback in [2, 4, 8]:
            trajectory = smoothen_trajectory(trajectory, group_size=16 // sessions_per_feedback)
        xs = np.array(range(len(trajectory))) * common_params['n_clients'] * sessions_per_feedback * 1e-3 
        ax[model_id].plot(xs, trajectory, label=f"{sessions_per_feedback} interactions per feedback")
        ax[model_id].set_title(model_name)
ax[2].legend(loc="lower right")
ax[0].set_ylabel("Mean MaxRR over batch")
ax[1].set_xlabel("# interactions, thousands")

plt.show()
f.savefig('n_interactions.png')
```
